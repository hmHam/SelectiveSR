$ G(x, y) = \frac{1}{2\pi \sigma} exp ( - \frac{x^2 + y^2}{2\sigma^2})$  
平均ベクトル$\mu = (0, 0)$, $X, Y$は独立で、標準偏差$\sigma_1, \sigma_2 = 1.3$  
の2変量正規分布を用いて、ガウスフィルタ作ってみた

```python
# 5x5のフィルタ
import numpy as np
from scipy.stats import multivariate_normal

Mu  = np.array([0, 0])
Sigma = np.array([
    [1.3**2, 0],
    [0, 1.3**2],
])

X, Y = np.meshgrid(
    np.arange(-2, 3, 1),
    np.arange(-2, 3, 1)
)
```

```python
f  = np.vectorize(lambda x, y: multivariate_normal(Mu, Sigma).pdf([x, y]))
Z = f(X, Y)
K = np.floor(Z*1000)
print(np.floor(K / K.sum() * 100))
K.sum()
36 / 256
```

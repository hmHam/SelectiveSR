```python
# 5x5のフィルタ
import numpy as np
from scipy.stats import multivariate_normal

Mu  = np.array([0, 0])
Sigma = np.array([
    [1.3, 0],
    [0, 1.3],
])

X, Y = np.meshgrid(
    np.arange(-2, 3, 1),
    np.arange(-2, 3, 1)
)
```

```python
X
```

```python
Y
```

```python
f  = np.vectorize(lambda x, y: multivariate_normal(Mu, Sigma).pdf([x, y]))
Z = f(X, Y)
Z
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1)
plt.show()
```
